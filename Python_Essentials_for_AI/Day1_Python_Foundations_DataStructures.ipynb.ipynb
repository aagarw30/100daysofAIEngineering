```python
# Strings
name = "Abhinav"
quote = 'Data drives intelligence'
print(type(name), type(quote))
```

```python
# Strings are indexable and sliceable
print(name[0])     # A
print(name[-1])    # v
print(name[1:4])   # bhi
print(name[:5])    # Abhin
print(name[3:])    # nav    
```

```python
# Boolean type - useful for conditional statements and logical operations.
is_python_easy = True
is_data_fun = False
print(type(is_python_easy))
print(is_python_easy and is_data_fun)  # False
print(is_python_easy or is_data_fun)   # True

```

```python
# Sequence Types - lists, tuples, ranges
# Lists - ordered, mutable collections of items. Useful for storing datasets, feature vectors.
fruits = ["apple", "banana", "cherry"]
print(type(fruits))

print(fruits[0])  # apple

fruits.append("date")

print(fruits)

fruits[0] = "apricot"
print(fruits)  # ['apricot', 'banana', 'cherry', 'date']


```

```python
# Tuples - ordered, immutable collections. Useful for fixed datasets, coordinates.
coordinates = (10.0, 20.0)
print(type(coordinates))


print(coordinates[1])  # 20.0
# coordinates[0] = 15.0  # This will raise an error

# Instead, create a new tuple
new_coordinates = (15.0, coordinates[1])
print(new_coordinates)  # (15.0, 20.0)
```

```python
# Ranges - represent sequences of numbers, often used in loops.

for i in range(1, 5):
    print(i)

num_range = range(0, 10, 2)
print(type(num_range))

for num in num_range:
    print(num)



```

```python
# Mapping Type - dictionaries - Great for structured data, configurations, or JSON handling.

student = {"name": "Alice", "age": 24, "major": "Data Science"}
print(type(student))

print(student["name"])  # Alice

student["age"] = 25

print(student)  # {'name': 'Alice', 'age': 25, 'major': 'Data Science'}

```

```python
# Set Types - sets, frozensets
# Sets - unordered collections of unique items. No duplicates. Commonly used for membership testing, removing duplicates, or set operations (union, intersection).

unique_numbers = {1, 2, 3, 2, 1}

print(type(unique_numbers))
print(unique_numbers)  # {1, 2, 3}

# Frozensets - immutable sets. Useful when you need a hashable set (e.g., as dictionary keys).
frozen = frozenset([1, 2, 3, 2, 1])

print(type(frozen))
print(frozen)  # frozenset({1, 2, 3})

# Demonstrating immutability
# frozen.add(4)  # This will raise an error 

```

```python
# None Type - represents the absence of a value. Useful for default values, optional parameters, or signaling 'no data'.

data = None
print(type(data))

print(data is None)  # True

```

```python
# Python Operators
a, b = 7, 3

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division (float)
print(a // b)  # Floor division
print(a ** b)  # Exponentiation
print(a % b)   # Modulus
```

```python
# Comparison Operators

print(a == b)  # Equal

print(a != b)  # Not equal

print(a > b)   # Greater than

print(a < b)   # Less than

print(a >= b)  # Greater than or equal to

print(a <= b)  # Less than or equal to

```

```python
# Logical Operators

print((a > b) and (a != b))  # Logical AND

print((a < b) or (a == b))   # Logical OR

print(not(a > b))             # Logical NOT

```

```python
# Membership Operators

fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)      # True

print("date" not in fruits)    # True

```

```python
# Identity Operators
x = [1, 2, 3]

y = x

print(x is y)      # True

print(x is not y)  # False

y = x.copy()

print(x is y)      # False

print(x == y)      # True

```

```python
# Assignment Operators

a += b  # a = a + b

print(a)  # 10

a *= b  # a = a * b

print(a)  # 30

a -= b  # a = a - b

print(a)  # 27

a /= b  # a = a / b

print(a)  # 9.0

a %= b  # a = a % b

print(a)  # 0.0

a **= b  # a = a ** b

print(a)  # 0.0

a //= b  # a = a // b

print(a)  # 0.0

```

```python
# String Operators

s1 = "Data"

s2 = "Science"

print(s1 + " " + s2)  # Concatenation

print(s1 * 3)          # Repetition

print("Sci" in s2)     # Membership

print(s1[0])           # Indexing

print(s2[1:4])        # Slicing

# String Methods

print(s1.lower())        # Lowercase

print(s2.upper())        # Uppercase

print(s1.replace("a", "o"))  # Replace

print(s2.split("e"))     # Split

print(s1.strip())        # Strip whitespace

# String Formatting

formatted = f"{s1} is fun!"

print(formatted)

# Using format() method

formatted2 = "{} makes {} interesting.".format(s1, s2)

print(formatted2)

# Using % operator

formatted3 = "%s and %s go well together." % (s1, s2)

print(formatted3)

# Escape Characters

print("Hello\nWorld")  # Newline

print("Tab\tSeparated")  # Tab

print("He said, \"Python is great!\"")  # Quotes

# Raw Strings

print(r"C:\Users\Admin\Desktop")  # Raw string to avoid escape sequences

# Triple Quotes for Multi-line Strings

multi_line = """This is a multi-line string.
It can span multiple lines.
Like this."""
print(multi_line)   

# String Methods Continued
print(s1.startswith("Da"))  # Starts with   

print(s2.endswith("ce"))    # Ends with

print(s1.count("a"))        # Count occurrences

print(s2.find("e"))         # Find substring index

print(len(s1))              # Length of string

print(s1.isalpha())         # Check if all characters are alphabetic

print(s2.isdigit())         # Check if all characters are digits

print(s1.capitalize())      # Capitalize first letter

print(s2.title())           # Title case

print(s1.center(20, '-'))   # Center with padding

# String Immutability Demonstration

original = "Hello"

modified = original.replace("H", "J")

print(original)  # Hello

print(modified)  # Jello
```

```python
# Type Casting Examples

# String to Integer
num_str = "42"
num_int = int(num_str)
print(type(num_int), num_int)  # <class 'int'> 42

# String to Float
num_str_float = "3.14"
num_float = float(num_str_float)
print(type(num_float), num_float)  # <class 'float'> 3.14

# Integer to String
num = 100
num_str2 = str(num)
print(type(num_str2), num_str2)  # <class 'str'> 100

# Float to Integer
float_num = 9.99
int_num = int(float_num)
print(type(int_num), int_num)  # <class 'int'> 9

# Integer to Float
int_num2 = 7
float_num2 = float(int_num2)
print(type(float_num2), float_num2)  # <class 'float'> 7.0

# Boolean to Integer
bool_val = True
int_from_bool = int(bool_val)
print(type(int_from_bool), int_from_bool)  # <class 'int'> 1

# Integer to Boolean
int_val = 0
bool_from_int = bool(int_val)
print(type(bool_from_int), bool_from_int)  # <class 'bool'> False

# Float to String
float_val = 2.718
str_from_float = str(float_val)
print(type(str_from_float), str_from_float)  # <class 'str'> 2.718
```

```python
# Dynamic Typing Demonstration - You don’t have to declare data types explicitly
var = 10
print(type(var), var)  # <class 'int'> 10

var = "Now I'm a string"
print(type(var), var)  # <class 'str'> Now I'm a string
```

Mini Practice Program -1 (Understanding User Input, Type Casting, f-strings formating, applying what is learned so far)

Task:
Write a program that takes two numbers as input and prints:
Their sum, difference, and product
Whether the first number is greater than, equal to, or less than the second
Whether both are even or odd
Use f-strings for the output formatting

```python
# Take first number from the user
num1 = float(input("Enter first number: ")) # type casting to float
print(num1)
```

```python
num2 = float(input("Enter second number: ")) # type casting to float 
print(num2)
```

```python
# Perform arithmetic operations
print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
print(f"Floor Division: {num1 // num2}")
print(f"Modulus: {num1 % num2}")
print(f"Exponentiation: {num1 ** num2}")
```

```python
# Print comparison results
print(f"{num1} == {num2}: {num1 == num2}")
print(f"{num1} != {num2}: {num1 != num2}")
print(f"{num1} > {num2}: {num1 > num2}")
print(f"{num1} < {num2}: {num1 < num2}")
print(f"{num1} >= {num2}: {num1 >= num2}")
print(f"{num1} <= {num2}: {num1 <= num2}")
```

```python
# Even or Odd Check
if int(num1) % 2 == 0:
    print(f"{int(num1)} is Even")
else:
    print(f"{int(num1)} is Odd")
```

```python
# Check if num2 is odd or even
if int(num2) % 2 == 0:
    print(f"{int(num2)} is Even")
else:
    print(f"{int(num2)} is Odd")
```

Mini Practice Program 2 - BMI Calculator - create a small program that uses user input, casting, and dynamic variables.

```python
name = input("Enter your name: ")
weight = float(input("Enter your weight (kg): "))
height = float(input("Enter your height (m): "))
# Print Name
print(f"Name: {name}")

# Print Weight
print(f"Weight: {weight} kg")

# Print Height
print(f"Height: {height} m")
```
