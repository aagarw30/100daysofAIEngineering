Loops â€” Repetition
Loops allow your program to repeat blocks of code automatically

```python
# for loop with list
# Print each fruit in the list
# The loop iterates over each item in the 'fruits' list
# For each iteration, the current item is assigned to the variable 'fruit'
# Then, it prints "I love" followed by the fruit name
# This continues until all items in the list have been processed
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("I love", fruit)
```

```python
# for loop with range
# Print numbers from 1 to 5
# Using range() function
# range(start, stop) generates numbers from start to stop-1
# Here, it generates numbers from 1 to 5
# So, the loop will iterate 5 times
for i in range(1, 6):
    print("Iteration:", i)
```

```python
# While Loop
# Print "Learning Python..." three times with count
# Initialize count to 0
# Continue looping while count is less than 3
# In each iteration, print the message and increment count by 1
# This continues until count reaches 3
count = 0

while count < 3:
    print("Learning Python...", count)
    count += 1
```

Loop Control Statements

| Command    | Description                |
| ---------- | -------------------------- |
| `break`    | Exits loop immediately     |
| `continue` | Skips current iteration    |
| `pass`     | Placeholder (does nothing) |


```python
# Example of continue statement - skip current iteration
# Explanation as follows: The loop will iterate from 1 to 5, but when the value of i reaches 3, the continue statement will skip the rest of the loop body for that iteration and move to the next iteration.
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# Example of break statement - Exit loop
# Example to demonstrate break
# Explanation as follows : The loop will iterate from 1 to 5, but when the value of i reaches 4, the loop will break and stop executing further.

for i in range(1, 6):
    if i == 4:
        print("Breaking the loop at", i)
        break
    print(i)
```

```python
# Example of pass statement - Placeholder - do nothing
# Explanation as follows: The loop will iterate from 1 to 5, but when the value of i reaches 3, the pass statement will do nothing and the loop will continue to the next iteration.
for i in range(1, 6):
    if i == 3:
        pass  # Placeholder, does nothing
    print(i)
```

Functions â€” Modular, Reusable Code

```python
# Just a simple function
# Explanation as follows: 
# This function takes a name as input and prints a greeting message.
# Important point to note for a function: Functions help in organizing code into reusable blocks.

def greet(name):
    print(f"Hello {name}, welcome to Day 2!")

greet("Abhinav")
```

```python
# Function with return
# Explanation as follows: 
# This function takes two numbers as input and returns their sum.
# Important point to note for a function with return: The return statement allows the function to send back a result to the caller.

def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Sum:", result)
```

```python
# Default arguments
# Explanation as follows: 
# This function has a default argument 'name' with the value "Guest".
# How it works: If no argument is provided when calling the function, it uses the default value. If an argument is provided, it overrides the default.

def greet_user(name="Guest"):
    print(f"Hello {name}!")
greet_user()  # Uses default argument
greet_user("Abhinav")  # Overrides default argument
```

Lambda (Anonymous Functions) - Short, inline functions
| Syntax                | Description                                 |
| --------------------- | ------------------------------------------- |

| `lambda args: expr`   | Creates a small anonymous function         |
| `map(func, iterable)` | Applies `func` to each item in `iterable`  |
| `filter(func, iterable)` | Filters `iterable` using `func`          |


```python
# A simple example
# How it works: This lambda function takes one argument x and returns its square. It is then called with the value 5, resulting in 25.
square = lambda x: x**2

print(square(5))
```

```python
# Where the lanbda function could be useful
# An example of data transformation using lambda with map
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
print(squared)

# How this works: The map function applies the lambda function (which squares each number) to each item in the nums list. The result is converted to a list and printed, showing the squared values.
```

```python
# Practice Exercise 1:
# Create a function that checks if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(11)  # True
is_prime(4)   # False
```

```python
# Practice Exercise 2:
# Create a function that returns the factorial of a number

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(5)  # 120
```

```python
# Practice Exercise 3:
# Create a function that reverses a string

def reverse_string(s):
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s   
reverse_string("Hello")  # "olleH"
```

```python
# Practice Exercise 4:
# Create a function that finds the largest number in a list


def find_largest(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest

find_largest([3, 5, 2, 8, 1])  # 8
```

```python
# Practice Exercise 5:
# Create a function that counts the number of vowels in a string

def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

count_vowels("Hello World")  # 3
```

```python
# Practice Exercise 6:
# Create a function that checks if a string is a palindrome
# go by basic what learned so far

def is_palindrome(s):
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return s == reversed_s

# check for palindrome
is_palindrome("radar")  # True
is_palindrome("hello")  # False
```

```python
# Create a number guessing game

import random

def number_guessing_game():
    number_to_guess = random.randint(1, 100)
    attempts = 0
    guessed = False

    print("Welcome to the Number Guessing Game!")
    print("I have selected a number between 1 and 100. Can you guess it?")

    while not guessed:
        user_guess = int(input("Enter your guess: "))
        attempts += 1

        if user_guess < number_to_guess:
            print("Too low! Try again.")
        elif user_guess > number_to_guess:
            print("Too high! Try again.")
        else:
            guessed = True
            print(f"Congratulations! You've guessed the number {number_to_guess} in {attempts} attempts.")

number_guessing_game()
```

```python
# Studen Score Analyzer program
def analyze_scores():
    num_students = int(input("Enter number of students: "))
    results = {}

    for _ in range(num_students):
        name = input("Enter student name: ")
        score = int(input("Enter score: "))

        if score >= 90:
            grade = "Excellent"
        elif score >= 75:
            grade = "Good"
        elif score >= 50:
            grade = "Average"
        else:
            grade = "Needs Improvement"

        results[name] = grade

    print("\n--- Report Card ---")
    for student, grade in results.items():
        print(f"{student} â†’ {grade}")

# Run the program
analyze_scores()
```
